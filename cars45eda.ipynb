```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv("car45_data.csv")
```

```python
# Preview the first few rows
print(df.head())
```

```python
# Check dataset information
print(df.info())
```

```python
# Summary statistics
print(df.describe())
```

```python
#DATA CLEANING
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Drop rows with missing values
df = df.dropna()
```

```python
# Check for duplicates
print("Duplicates: ", df.duplicated().sum())
```

```python
# Verify data types
print(df.dtypes)
```

```python
#EXPLORATORY DATA ANALYSIS(EDA)
```

```python
import matplotlib.pyplot as plt
```

```python
# Group by make and amount sales
sales_trend = df.groupby('make')['amount'].sum()
```

```python
# Plot sales trend
plt.figure(figsize=(5, 4))
sales_trend.plot()
plt.title("Sales Over Time")
plt.xlabel("MAKE")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

```python
# Sales by Region
region_sales = df.groupby('region')['amount'].sum()

# Plot sales by region
region_sales.plot(kind='bar', figsize=(20, 8), color='skyblue')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
# Top 10 products by sales
top_products = df.groupby('make')['amount'].sum().sort_values(ascending=False).head(10)

# Plot top products
top_products.plot(kind='barh', figsize=(8, 5), color='orange')
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Total Sales")
plt.ylabel("Product")
plt.show()
```

```python
#CORRELATION ANALYSIS
```

```python
import seaborn as sns
```

```python
# Encode categorical variables ('region' and 'make') into numerical form
df['region_encoded'] = df['region'].astype('category').cat.codes
df['make_encoded'] = df['make'].astype('category').cat.codes
```

```python
# Select columns for correlation
correlation_columns = ['region_encoded', 'amount', 'make_encoded']
```

```python
# Calculate the correlation matrix
correlation_matrix = df[correlation_columns].corr()
```

```python
# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', xticklabels=['Region', 'Amount', 'Make'], yticklabels=['Region', 'Amount', 'Make'])
plt.title("Correlation Heatmap")
```

```python
#KEY FINDINGS 
#1. Pontiac had the highest amount for sales
#2.Lagos state, Ikeja had the highest sales by region and Abuja(FCT), Wuse had the lowest sales by region
#3.Toyota was number 1 best selling product 
```
